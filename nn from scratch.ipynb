```python
import numpy as np
x1 = 0.5
x2 = 0.8
weights = [0.40, 0.45, 0.67, 0.21]
biases = [0.8, 0.6]

z1 = x1*weights[0] + x2*weights[1] + biases[0]
z2 = x1*weights[2] + x2*weights[3] + biases[1]

print(f"hidden layer 1st neuron weighted sum for input x1: {z1} \nhidden layer 2nd neuron weighted sum for input x2: {z2}")

print("\n")

def sigmoid(z):
    return 1/(1 + np.exp(-z))

a1 = sigmoid(z1)
a2 = sigmoid(z2)

print(f"output from the hidden layer: neuron1 {a1}")
print(f"output from the hidden later: neuron2 {a2}")

op_weight = [0.57, 0.78]
op_bias = [0.45]


h1 = a1*op_weight[0] + a2*op_weight[1] + op_bias[0]

o1 = sigmoid(h1)

print(f'output from the output layer: {o1}')
```

```python
import numpy as np 

no_input = 2
no_of_layers = 3 #no of hidden layers
neurons = [2, 2, 2] # no of neurons per layer of the hidden layer 
output = 1 # no of nodes in the output layer 
```

```python
def network_initiate(no_input, no_of_layers, neurons, output):
        
    network = {}
    
    no_of_previous_nodes = no_input
    
    for layer in range(no_of_layers + 1): 
        if layer == (no_of_layers):
            layer_name = 'output_layer'
            num_of_nodes = output
        else: 
            layer_name = f'layer{layer + 1}'
            num_of_nodes = neurons[layer]
            
        network[layer_name] = {}
        
        for node in range(num_of_nodes): 
            #initializing weights and bias for each neuron 
            node_name = f"node{node + 1}"
            network[layer_name][node_name] = {
                'weight' : np.random.uniform(size = no_of_previous_nodes),
                'bias' : np.random.uniform(size = 1)
            }
            no_of_previous_nodes = neurons #Each neuron needs one weight per neuron in the previous layer. So the next layer needs to know how many neurons the current layer has.
    
    
    return network
```

```python
def weighted_sum(inputs, weight, bias):
    return np.sum(weight * inputs) + bias

def activation(weighted_sum):
    return 1/(1 + np.exp(-weighted_sum))
def input_gen(no_input):
    return np.random.uniform(size = no_input)
```

```python
#fully connected neural network 
my_net = network_initiate(2, 3, [2, 2, 2], 1)
inp = input_gen(2)
x = foreward_prop(my_net, inp)
print(x)
```
